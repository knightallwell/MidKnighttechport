```python
price_df= pd.read_csv(r'C:\Users\allwe\Downloads\price_data (1).csv')
```

```python
price_df.head()
```

```python
price_df['price_date']=pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')
```

```python
price_df
```

```python
monthly=price_df.groupby(['id','price_date']).agg({'price_off_peak_var':'mean','price_off_peak_fix':'mean'}).reset_index()
```

```python
jan_price_bev=monthly.groupby('id').first().reset_index()
```

```python
jan_price_bev
```

```python
dec_price_bev=monthly.groupby('id').last().reset_index()
```

```python
dec_price_bev
```

```python
Dec_rename=dec_price_bev.rename(columns=({'price_off_peak_var': 'dec_1', 'price_off_peak_fix':'dec_2'}))
```

```python
Dec_rename
```

```python
jan_dropped=jan_price_bev.drop(columns=['price_date'])
```

```python
jan_dropped
```

```python
merge_diff=pd.merge(Dec_rename,jan_dropped, on='id')
```

```python
merge_diff
```

```python
merge_diff['offpeak_diff_dec_january_energy'] = merge_diff['dec_1'] - merge_diff['price_off_peak_var']
```

```python
merge_diff['offpeak_diff_dec_january_power'] = merge_diff['dec_2'] - merge_diff['price_off_peak_fix']
```

```python
merge_diff = merge_diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
```
