```python
def extract_year(date_list):
    years = []
    for date in date_list:
        years.append(date.split("-")[0])
    return years
```

```python
# Assuming visibm['Date'] is a list of date strings
```

```python
date_list = visibm['Date']
```

```python
year = extract_year(date_list)
```
